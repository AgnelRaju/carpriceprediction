```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

### Hyperparameter tuning

```python
import numpy as np
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
```

```python
print(n_estimators)
```

```python
# params 
min_samples_split=[2,5,10,15,100] # minimum number of samples required to split a node
min_samples_leaf=[1,2,5,10]
max_features=['auto','sqrt']
max_depth = [int(x) for x in np.linspace(start=5,stop=30,num=6)]# minimum number of levels in a tree
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# creating the ramdom grid
random_grid = {"n_estimators":n_estimators,
               "max_features":max_features,
               "max_depth":max_depth,
               "min_samples_split":min_samples_split,
               "min_samples_leaf":min_samples_leaf}
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring="neg_mean_squared_error",n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
# the below one is my model
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred = rf_random.predict(X_test)
```

```python
mean_squared_error(Y_test,y_pred)
```

```python
sns.distplot(Y_test-y_pred)# the differnce beteween the y_pred and y_test should be minimum which say the accuracy is good scorer
```

```python
plt.scatter(Y_test,y_pred)
```
